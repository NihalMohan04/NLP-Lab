```python
from transformers import pipeline
classifer = pipeline("sentiment-analysis")
```

```python
classifer("We are very happy to wshow you the Trans lib")
```

```python
classifer("The pizza is not good at all, i didnt like it")
```

```python
classifer("pizza taste is not good but it looks awesome")
```

```python
classifer("The pizza is not that great but the crust is awesome")
```

```python
classifer("The pizza is very bad")
```

```python
results=classifer(["We are very happy here.","We hope you are doing great.", "I guess you won't hate it.","I hope you will love to join NLP class."])

for result in results:
  print(f"Label:{result['label']}, with socre:{round(result['score'],4)}")
```
